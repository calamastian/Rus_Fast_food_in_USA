```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df_1 = pd.read_csv('Datafiniti_Fast_Food_Restaurants_May19.csv')
df_1.head()
```

```python
df_2 = pd.read_csv('Datafiniti_Fast_Food_Restaurants.csv')
df_2.head()
```

```python
df_3 = pd.read_csv('FastFoodRestaurants.csv')
df_3.head()
```

```python
#очистим данные от лишней информации
df_1 = df_1.drop(['id', 'keys', 'sourceURLs', 'websites', 'postalCode', 'dateAdded', 'dateUpdated', 'primaryCategories'], axis=1)
df_2 = df_2.drop(['id', 'keys', 'sourceURLs', 'websites', 'postalCode', 'dateAdded', 'dateUpdated'], axis=1)
df_3 = df_3.drop(['keys', 'websites', 'postalCode'], axis=1)
```

```python
df_1.head()
```

Приведем датафреймы к одному виду:
* df_1 address	categories	city	country	latitude	longitude	name	province
* df_2 address	categories  city	country	latitude	longitude	name	province
* df_3 address	categories  city	country	latitude	longitude	name	province

```python
# заменим пропущеный столбец значением Unknown
df_3.insert(1, 'categories', 'Unknown')
df_3.head()
```

```python
#Объединим 3 датафрейма в 1 для удобства
df_svod = pd.concat([df_1, df_2, df_3])
df_svod.head()
```

```python
#Оставим только города, которые мы упоминали в юнит-экономике
df = df_svod[df_svod['city'].isin(['Columbus', 'Portland', 'Cleveland', 'Cincinnati','Indianapolis', 'Birmingham', 'Miami', 'Las Vegas'])]
provinces = ['OH', 'NV', 'IN', 'FL', 'OR', 'AL']
df = df[df.province.isin(provinces)]
df = df[~((df['city'] == 'Portland') & (df['province'] != 'OR'))]
df = df[~((df['city'] == 'Columbus') & (df['province'] != 'OH'))]
```

```python
#приведем похожие названия к одному виду
most_similar = [["Arbys", "Arby's"],
['Burger King', 'Burger King - Temporarily Closed'], 
['Burgerville', 'Burgerville, USA'],
['Carls Jr', "Carl's Jr."], 
['Checkers', "Checker's"],
['ChickfilA', 'Chick-fil-A'],
['Churchs Chicken', "Church's Chicken"],
['Dominos Pizza', "Domino's Pizza"], 
['Dunkin Donuts', "Dunkin' Donuts"],
['Fazolis', "Fazoli's"],
['Jersey Mikes Subs', "Jersey Mike's Subs"],
['Jimmy Johns', "Jimmy John's"],
['Kfc','Kentucky Fried Chicken', 'KFC - Closed'],
['McAlisters Deli', "McAlister's Deli"],
['McDonalds', "McDonald's"],
['Moes Southwest Grill', "Moe's Southwest Grill"], 
['Mr Hero', 'Mr. Hero'], 
['Mrs Fields', "Mrs. Fields"],
['Nicholas Restaurant', "Nicholas' Restaurant"],
['Panda Express', 'Panda Express - Inside Vons + Inside Vons +'],
['Popeyes Louisiana Kitchen', "Popeye's Louisiana Kitchen"],
['Rallys', "Rally's"], 
['SONIC Drive In', 'Sonic Drive-In'],
['Subway', 'SUBWAY¬Æ'],
["Wendy's", 'Wendys']]
```

```python
name_dict = {}
for row in most_similar:
    for similar_word in row:
        name_dict[similar_word] = row[0]
name_dict
```

```python
#заменим названия ресторанов в датафрейме
names = df['name'].values
for i in range(len(names)):
    if name_dict.get(names[i]) != None:
        names[i] = name_dict[names[i]]
df['names'] = names
df.names
```

```python
plt.savefig('graph.png')
```

```python
import folium
USA_map = folium.Map(
    location = [40.827082, -101.472679],    # широта и долгота 
    zoom_start = 4.5)
lat  = df['latitude']
lon = df['longitude']
name = df['names']

for lat, lon, name in zip(lat, lon, name):
    marker = folium.CircleMarker(
        location=[lat, lon],
        fill_opacity=0.6,
        radius=7,
        popup=name  #всплывающее окно
    )
    marker.add_to(USA_map) #маркеры
# Отобразить карту 
USA_map
```
