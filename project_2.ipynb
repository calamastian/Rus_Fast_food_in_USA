Ключевые гипотезы:
* Прибыль зависит от плотности населения и количества конкурентов
* Молодое население (18-35 лет) чаще посещает фастфуд
* Более высокий доход населения → больше трат на общепит
* Высокий уровень преступности негативно влияет на посещаемость

Введем данные из открытого источника и юнит-экономики по нужным нам городам в новый датафрейм:

```python
data = {
    'city': ['Cincinnati', 'Las Vegas', 'Columbus', 'Miami', 'Indianapolis', 
             'Cleveland', 'Portland', 'Birmingham'],
    'counts': [205, 200, 166, 163, 161, 109, 108, 81],  # количество конкурентов
    'square': [206.3, 352, 586, 143.1, 953, 213.6, 375.6, 387.3],  # площадь, км²
    'population': [311097, 678922, 913175, 487014, 879293, 362656, 630498, 196644],
    'percent_18_35': [32.40, 22.50, 25.70, 25.30, 26.40, 27.00, 26.60, 29.90], #доля населения от 18-35 лет
    'median_income': [51707, 70723, 78538, 59390, 62995, 39187, 88792, 44376],  # $
    'order_profit': [3.07, -0.50, 1.90, 0.55, 1.96, 3.62, 4.90, 1.28],  # предположим: средняя прибыль с заказа, $
    'crime_rate': [50.06, 31.74, 31.52, 32.79, 46.56, 64.35, 66.99, 62.65] #уровень преступности
}
df_1 = pd.DataFrame(data)
df_1.info()
```

```python
df_1['population_density'] = df_1['population'] / df_1['square']  # чел/км²
df_1['competition_density'] = df_1['counts'] / df_1['square']  # концентрация конкурентов/км²
df_1['target_audience'] = df_1['population'] * (df_1['percent_18_35'] / 100)  # целевая аудитория 18-35 лет
df_1['potential_demand'] = df_1['target_audience'] * (df_1['median_income'] / 10000)  # условный показатель спроса
df_1
```

```python
#Нормализируем значения
from sklearn.preprocessing import MinMaxScaler
metrics_to_normalize = ['population_density', 'competition_density', 
                       'median_income', 'crime_rate', 'target_audience',
                       'potential_demand']
scaler = MinMaxScaler()
df_normalized = df_1.copy()
df_normalized[metrics_to_normalize] = scaler.fit_transform(df_1[metrics_to_normalize])
df_normalized
```

```python
weights = {
    'population_density': 0.30,    # умеренная важность
    'competition_density': -0.60,  # важный негативный фактор
    'median_income': 0.70,         # самый важный фактор
    'crime_rate': -0.20,           # умеренный негатив
    'target_audience': 0.30,       # умеренная важность
    'potential_demand': 0.50       # высокая важность
}
df_1['score'] = (
    df_normalized['population_density'] * weights['population_density'] +
    df_normalized['competition_density'] * weights['competition_density'] +
    df_normalized['median_income'] * weights['median_income'] +
    df_normalized['crime_rate'] * weights['crime_rate'] +
    df_normalized['target_audience'] * weights['target_audience'] +
    df_normalized['potential_demand'] * weights['potential_demand']
)
df_sorted = df_1.sort_values('score', ascending=False)
print("Рейтинг городов по привлекательности:")
display(df_sorted[['city', 'attractiveness_score', 'population', 'median_income', 'competition_density']])
```
